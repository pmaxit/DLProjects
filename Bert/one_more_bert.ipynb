# Transformer

First step: take the embedding input 256 dimension and split into 8 parts.

Now, 8 of these parts all have 32 dimension. 

They will be sent to linear layer and form query, key , value for self attention.

Scaled dot product attention

```python
import torch
import torch.nn as nn
```

```python
class SelfAttention(nn.Module):
    def __init__(self, embed_size: int, heads: int):
        ''' splitting the embed_size in heads part. Self attention '''
        super(SelfAttention, self).__init__()
        self.embed_size = embed_size
        self.heads = heads
        self.head_dim = embed_size // heads
        
        # dimension for each head
        assert (self.head_dim*heads == embed_size), "Embed size needs to be divisible by heads"
        
        self.values = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.keys = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.queries = nn.Linear(self.head_dim, self.head_dim, bias=False)
        
        
        
        
        
        
```
